### CADERNO DE ANÁLISE EXPLORATÓRIA DE DADOS DO DATASUS (PROCESSAMENTO)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
dados = pd.read_csv("csvs\\datasus_local_internacao_processamento.csv", encoding = "utf-8", skiprows = 3, sep = ";", 
                    skipfooter = 10, engine = 'python', decimal=",", na_values='-')
```

```python
dados.head() #mostra as primeiras 5 linhas (header), para as últimas 5 utilizar .tail()
```

Formatar os valores para 2 casas decimais sem uso de exponenciação

```python
pd.options.display.float_format = "{:.2f}".format 
```

Para selecionar uma coluna, utilizamos square brackets com a key como em dicts

```python
dados['2008/Jan'].mean()
```

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind='bar', figsize=(8,5))


formatter = ticker.StrMethodFormatter('{x:,.2f}')
ax.yaxis.set_major_formatter(formatter)
ax.set_xlabel('')

#DESAFIO 1: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.
plt.grid()
plt.title("Valor Gasto (em R$) em Procedimentos Hospitalares do SUS por UF em Agosto de 2008")
```
